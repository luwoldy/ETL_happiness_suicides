```python
#Create dataframe with suicide data
import pandas as pd
import numpy as np
file = 'Data/suicides.csv'
suicides_df = pd.read_csv(file)[['country','year','suicides_no']]
suicides_df_2015 = suicides_df.loc[suicides_df['year'] == 2015]
suicides_df_2015
```

```python
#Group by country and year and get the total number of suicides
country_group = suicides_df_2015.groupby(['country','year']).sum().reset_index()[['country','suicides_no']]
country_group
```

```python
#Import the happiness sqllite
import sqlite3
# Create your connection.
cnx = sqlite3.connect('Data/happiness.sqlite')
happiness_df = pd.read_sql_query("SELECT * FROM `2015`", cnx)[['Country','Happiness Score']]
```

```python
merged_df = happiness_df.merge(country_group, left_on='Country', right_on='country')
final_df = merged_df[['Country','Happiness Score','suicides_no']]
final_df.sort_values(['suicides_no'],ascending=False)
```

```python
# Upload new DF into Mongo
import pymongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
db = client.CountriesDB
collection = db.items


for index, row in final_df.iterrows():
    post = {
        "Country": row['Country'],
        "Happiness Score": row['Happiness Score'],
        "Number of Suicides": row['suicides_no']
    }
    collection.insert_one(post)
```

```python
# Verify results:
results = collection.find()
for result in results:
    print(result)
```
